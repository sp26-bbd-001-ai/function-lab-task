```python
#Q1. Simple Function
from datetime import datetime

def greet_with_time(name):
    current_time = datetime.now().strftime("%H:%M:%S")
    print(f"Hello, {name}! Current time is {current_time}")

greet_with_time("Ali")
```

```python
#Q2. Function with Multiple Parameters
def calculate_average(num1, num2, num3):
    return (num1 + num2 + num3) / 3

print(calculate_average(10, 20, 30))
```

```python
#Q3. Default Argument
def calculate_discount(price, discount=10):
    final_price = price - (price * discount / 100)
    return final_price

print(calculate_discount(1000))
print(calculate_discount(1000, 20))
```

```python
#Q4. Keyword Arguments
def print_info(name, age, city):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"City: {city}")

print_info(name="Ahmed", age=22, city="Lahore")

```

```python
#Q5. Variable-Length Arguments
def find_max(*numbers):
    return max(numbers)

print(find_max(3, 7, 2, 9, 5))
```

```python
#Q6. Recursive Function
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```python
#Q7. Function as an Argument
def apply_function(func, numbers):
    return [func(num) for num in numbers]

def double(x):
    return x * 2

print(apply_function(double, [1, 2, 3, 4]))
```

```python
#Q10. Function Decorators
import time

def timer_decorator(func):
    def wrapper():
        start_time = time.time()
        func()
        end_time = time.time()
        print(f"Execution Time: {end_time - start_time:.6f} seconds")
    return wrapper

@timer_decorator
def example_function():
    for i in range(1000000):
        pass

example_function()
```
